```python
import matplotlib.pyplot as plt

# Assigning values
numNoBuyLVT = 6438.0
numNoBuyDTI = 1871.0
numNoBuyFEDTI = 5586.0
numNoBuyCreditScore = 4028.0
numNoBuyMortgagePortionHigh = 5586.0

# Data for the pie chart
labels = ['No Buy (LVT)', 'No Buy (DTI)', 'No Buy (FEDTI)', 'No Buy (Credit Score)', 'No Buy (Mortgage Portion High)']
sizes = [numNoBuyLVT, numNoBuyDTI, numNoBuyFEDTI, numNoBuyCreditScore, numNoBuyMortgagePortionHigh]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightpink']
explode = (0.1, 0, 0, 0, 0)  # Explode the 1st slice (No Buy (LVT))

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Reasons for Not Buying a Home')
plt.show()
```
